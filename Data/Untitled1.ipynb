```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df_car = pd.read_csv('data\Vehicles.csv')
df_car.fillna(0)

df_bus = pd.read_csv('data\Bus.csv')
df_bus.fillna(0)

df_tram = pd.read_csv('data\Tram.csv')
df_tram.fillna(0)


time = df_car['Time'].to_numpy()
for i in range(len(time)):
    time[i] = time[i][2:]
    
def arrange_data1(df):
    #Extracts data from the dataframe and seperate
    DirText = []
    for i in range(14):
        DirText.append('Direction ' + str(i+1)) 

    directions = []

    for i in DirText:
        temp = (df[i].to_numpy())
        directions.append(temp)
        
    return directions


def arrange_data2(df_car,df_bus,df_tram,dirr):
    #Arranges specific to a single direction
    
    directions_car = arrange_data1(df_car)
    directions_bus = arrange_data1(df_bus)
    directions_tram = arrange_data1(df_tram)

    df = pd.DataFrame({'Time' : time})
    df['Car'] = df_car['Direction ' + dirr]
    df['Bus'] = df_bus['Direction ' + dirr]
    df['Tram'] = df_tram['Direction ' + dirr]

    return df

df.plot.bar(x= "Time", y=["Car" ,"Bus", "Tram" ], figsize=(16,5), stacked="True", xlabel="Time in Minutes", ylabel="Frequency")

```
